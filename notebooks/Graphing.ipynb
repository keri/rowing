# Features explained

- TimeSecondMean = mean for all boats in NumberType: Double,Single,Quad,Four,eight
- RowerXRowNumber = number of rows for that rower that day (i.e. second row for rower X)
- CombinedNumberRows = total number of rows for all rowers in the boat for that day (i.e. all rowers had a total of x rows before that race
- crew = names of rowers in boat in order of 1-8 position

# All Rowers in Each Category, time (bar=time in seconds) and (line=mean time for that type)

```python
types = ['single','double','quad','four']
df_types = [single,double,quad,four]
```

```python
for i, df in enumerate(df_types):
    _, ax = plt.subplots()
    t = types[i]
    df.sort_values(by=['TimeSeconds'],inplace=True)
    df.plot(x='crew', y='TimeSeconds', kind='bar',rot=90, ax=ax, colormap='coolwarm', figsize=(10,10))
    df.plot(x='crew', y='TimeSecondMean', rot=90, ax=ax, secondary_y=False, color='black',title=t)
    _.savefig(f'../data/{t}.png',bbox_inches='tight')
```
